```python
# imports
from astropy.utils.data import get_pkg_data_filename
from astropy.io import fits
import aplpy
import matplotlib.pyplot as plt
```

## *Example 1: Plotting slices of a 3D FITS cube with subplots*

```python
# making subplots of slices from FITS cube

co_cube = get_pkg_data_filename('WFPC2u5780205r_c0fx.fits')
fig = plt.figure(figsize=(11,9.1))
f1 = aplpy.FITSFigure('WFPC2u5780205r_c0fx.fits',figure=fig, subplot=[0,0.6,0.4,0.5], slices=[0], dimensions=[0,1])
f1.show_colorscale()
f1.add_grid()
f1.set_title('Slice 1')
f1.add_colorbar()
f1.colorbar.set_axis_label_text('counts/sec')
f1.colorbar.set_pad(0.15)

f2 = aplpy.FITSFigure('WFPC2u5780205r_c0fx.fits',figure=fig, subplot=[0.6,0.6,0.4,0.5], slices=[1], dimensions=[0,1])
f2.show_colorscale()
f2.add_grid()
f2.set_title('Slice 2')
f2.add_colorbar()
f2.colorbar.set_axis_label_text('counts/sec')
f2.colorbar.set_pad(0.15)

f3 = aplpy.FITSFigure('WFPC2u5780205r_c0fx.fits',figure=fig, subplot=[0,0,0.4,0.5], slices=[2], dimensions=[0,1])
f3.show_colorscale()
f3.add_grid()
f3.set_title('Slice 3')
f3.add_colorbar()
f3.colorbar.set_axis_label_text('counts/sec')
f3.colorbar.set_pad(0.15)

f4 = aplpy.FITSFigure('WFPC2u5780205r_c0fx.fits',figure=fig, subplot=[0.6,0,0.4,0.5], slices=[3], dimensions=[0,1])
f4.show_colorscale()
f4.add_grid()
f4.set_title('Slice 4')
f4.add_colorbar()
f4.colorbar.set_axis_label_text('counts/sec')
f4.colorbar.set_pad(0.15)

fig.suptitle('WFPC2 CCD Images of NGC4151',x=0.5,y=1.15)
```

## *Example 2: Plotting 2D FITS file with contours*

```python
gc = aplpy.FITSFigure('2MASS_k.fits')
#gc.show_grayscale()
#gc.show_colorscale()
#gc.show_colorscale(cmap='gist_heat')
gc.show_rgb('2MASS_arcsinh_color.png')
gc.tick_labels.set_font(size='small')
gc.show_contour('mips_24micron.fits', colors='white')
gc.add_grid()
#gc.remove_grid()

import numpy
ra, dec = numpy.loadtxt('yso_wcs_only.txt', unpack=True)
gc.show_markers(ra, dec, edgecolor='green', facecolor='none',
                marker='o', s=10, alpha=0.5)

gc.list_layers()

gc.show_markers(ra, dec, layer='marker_set_1', edgecolor='red',
                facecolor='none', marker='o', s=10, alpha=0.5)
gc.save('myfirstplot.png')
```
